### Class and Object

```python
class Phone:
    category = "Electronics"

    #Constructor
    def __init__ (self, model, battery, camera, battery_per = 100):
        self.model = model
        self.battery = battery
        self.camera = camera
        self.battery_per = battery_per
    #function declare
    def use_hour(self, hour):
        self.battery_per -= hour
    def photo(self, click):
        self.battery_per -= click
```

```python
apple = Phone("Iphone", 4000, 20)
vivo = Phone("Vivo", 3000, 30)

vivo.category = "Super Electronics"

print(apple.category)
print(vivo.category)

print(apple.model)
print(apple.battery_per)

apple.photo(10)

print(apple.battery_per)
apple.use_hour(2)
print(apple.battery_per)
```

### Inheritance

```python
class SmartPhone(Phone):
    def __init__(self, model, battery, camera, processor, battery_per=100):
        super().__init__(model, battery, camera, battery_per)
        self.processor = processor
```

```python
oppo = SmartPhone("Electronics", 4000, 30, "SD")

print(oppo.processor)
print(oppo.battery_per)
oppo.photo(3)
print(oppo.battery_per)
```

### Multiple Inheritance

```python
class CoolingSystem:
    def __init__(self, colling):
        self.colling = colling

class Android(SmartPhone, CoolingSystem):
    def __init__(self, model, battery, camera, processor, colling, battery_per=100):
        SmartPhone.__init__(self, model, battery, camera, processor, battery_per)
        CoolingSystem.__init__(self, colling)
```

```python
mi = Android("Electronincs", 3000, 20, "SD", "Nitrogen")

print(mi.colling)
mi.use_hour(10)
print(mi.battery_per)
```

## Polymorphisom

```python
class Camera():
    def __init__ (self, name):
        self.name = name

    def capture (self):
        print("photo is captured!")
```

```python
class Phonee(Camera):
    def __init__(self, name, resulation):
        Camera.__init__(self, name)
        self.resulation = resulation
    def capture(self):
        print("photo is captured by phone")

class DSLR(Camera):
    def __init__(self, name, resulation):
        Camera.__init__(self, name)
        self.resulation = resulation
    def capture(self):
        print("photo is captured by dslr")

class Drone(Camera):
    def __init__(self, name, resulation):
        Camera.__init__(self, name)
        self.resulation = resulation
    def capture(self):
        print("photo is captured by drone")
```

```python
p1 = Phonee("Iphone", 40)
d1 = DSLR("Canon", 200)
d2 = Drone("Nikkon", 150)

p1.capture()
```

## Encapsulation

```python
class Mobile():
    def __init__(self, name, model, iemi):
        self.name = name
        self.model = model
        self.__iemi = iemi

    def iemi_getter(self):
        return self.__iemi

    def iemi_setter(self, iemi):
        self.__iemi = iemi
        
pn = Mobile("Iphone", "17ProMax", "1eefx2")

print(pn.model)
print(pn.iemi_getter())
pn.iemi_setter("ahaha")
print(pn.iemi_getter())
```

## Abstraction
