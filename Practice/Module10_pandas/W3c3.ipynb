## List touple and dictonary to make a pandas dataframe

```python
lst = [['tan', 10], ['man', 20]]

ds = pd.DataFrame(lst, columns=['Name','Age']) # list theke pnadas dataframe

ds
```

```python
tpl = (('tan',10), ('man', 10))

ds = pd.DataFrame(tpl, columns=['Name', 'Age']) #touple to pandas dataframe conversion

ds
```

```python
dic = {
    'Name' : ['tan', 'man'],
    'Age' : [10,20]
}

ds = pd.DataFrame(dic) #dictionary to pandas dataframe

ds
```

```python
dic = [
    {'Name' : 'tan', 'Age' : 10, 'City' : 'Barishal'},
    {'Name' : 'man', 'Age' : 10, 'City' : 'Khulna'},
    {'Name' : 'tan', 'City' : 'Barishal'}
]

print(type(dic))

ds = pd.DataFrame(dic) #eikhane kintu null value show kortice 
# uporer way tea null value daoar kono option e chilo nah
ds
```
