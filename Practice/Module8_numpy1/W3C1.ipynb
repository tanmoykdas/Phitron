## Basic array operation

```python
arr = [1,2,3,4,5]
arr = (arr * 5)
print(arr) #eikhane sea single element dhore operation chalaite pare ni 

arr = [1,2,3,4,5]
arr = [x * 5 for x in arr] #eikhane parche kintu puro bisoyta complex
print (arr)
```

## Using Numpy (Numerical python)


```python
import numpy as np
arr = np.array([1,2,3,4,5])
arr = arr * 5
print(arr)
```

## ND array

```python
import numpy as np
arr1 = np.array([1,2,3])
print(type(arr1))
print(arr1.ndim)


arr2 = np.array([[1,2,3],[4,5,6]])
print(arr2.ndim)

arr3 = np.array (
    [[[1,2,3],[4,5,6]],
    [[1,2,3],[4,5,6]]]
)
print(arr3.ndim)
```

```python
#printing dimension
print(arr1.ndim)
print(arr2.ndim)
print(arr3.ndim)

#printing shape
print(arr1.shape)
print(arr2.shape)
print(arr3.shape)

#priting size
print(arr1.size)
print(arr2.size)
print(arr3.size)

#printing datatype
print(arr1.dtype)
print(arr2.dtype)
print(arr3.dtype)
```

## Random number generation using numpy

```python
#just calling rand

arr1 = np.random.rand(2,3)

print(arr1)
print(arr1.ndim)
print(arr1.size)
print(arr1.shape)
print(arr1.dtype)
```

```python
# Random number but int

arr1 = np.random.randint(1,5, (2,3))

print(arr1)
print(arr1.ndim)
print(arr1.size)
print(arr1.shape)
print(arr1.dtype)
```

```python
# Random number but float

arr1 = np.random.uniform(2, 5, (2, 4))

print(arr1)
print(arr1.ndim)
print(arr1.size)
print(arr1.shape)
print(arr1.dtype)
```

## Numpy creation using range

```python
# arange(start, end, step)

arr1 = np.arange(1, 10, 1)
print(arr1)
arr1 = arr1.reshape(3,3)
print(arr1)
```

```python
#linspace(start, end, koto gulo point dorkar)

arr1 = np.linspace(1, 10, 11)
print(arr1)
```

```python
#logspace(start, end, kotogulo dorkar)

arr1 = np.logspace(0, 4, 4) #bydefault 10 base ney jodi ami base nah dei 
print(arr1)
arr1 = np.logspace(0, 4, 4, base=2) 
print(arr1)
```

## Indexing and Slicing of a numpy array

```python
arr1 = np.array([[1,2,3],[4,4,6]])
arr2 = np.array([1,2,3,4])

#Slicing 1D array
row = arr2[1:4]
print(row)
print()

#slicing 2D array
row0 = arr1[0:1,]
row1 = arr1[1: , ]

print(row0)
print(row1)

col0 = arr1[::, 0:1]
col1 = arr1[::, 1:2]
print(col0)
print(col1)
```

## copy, Advance indexing, Iteration
