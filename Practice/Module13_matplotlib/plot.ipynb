### Using legend()

```python
programming_enrollment = df["Programming"]
digital_enrollment = df["Digital Marketing"]
year = df["Year"]

plt.title("Enrollment over year")
plt.xlabel("Year")
plt.ylabel("Enrollment")

plt.plot(year, programming_enrollment, label = "Programmin")
plt.plot(year, digital_enrollment, label = "digital marketing")

plt.legend()
```

### better ploting

```python
programming_enrollment = df["Programming"]
digital_enrollment = df["Digital Marketing"]
year = df["Year"]

plt.title("Enrollment over year")
plt.xlabel("Year")
plt.ylabel("Enrollment")

plt.plot(year, programming_enrollment, label = "Programmin", color='red', linewidth=3, linestyle="dashed",marker='o', markersize=5)
plt.plot(year, digital_enrollment, label = "digital marketing", color='green',linewidth=3,linestyle='dotted')


plt.legend()
plt.grid()
plt.show()
```

### Bar Chart

```python
data = {
    'Name': ['a', 'b', 'c', 'd', 'e'],
    'Gender': ['Male', 'Female', 'Male', 'Female', 'Male']
}

dataset = pd.DataFrame(data)

dataset
```

```python
gender_group = dataset.groupby(dataset["Gender"]).size()


plt.title("Male/Female")
plt.xlabel("Gender")
plt.ylabel("Count")

plt.bar(gender_group.index, gender_group.values, edgecolor='black', color=['red','yellow'])

```

```python
df = pd.read_csv("/home/tanmoy/Desktop/Phitron/Practice/Module13_matplotlib/student_data.csv")
df
```

```python
complition = df.groupby(df["CompletionStatus"]).size()

plt.title("Complition status")
plt.xlabel("Complition")
plt.ylabel("Count")
plt.bar(complition.index, complition.values, edgecolor='black', color=['red', 'orange', 'yellow'])
```
