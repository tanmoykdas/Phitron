```python
# we can change a column
import numpy as np
df["Python Marks"] = df["Python Marks"].fillna(df["Python Marks"].mean())
df["Python Marks"] = df["Python Marks"].round(1) # takes one value like 88.7

df
```

```python
df["Data Structure Marks"] = df["Data Structure Marks"].fillna(df["Data Structure Marks"].mean(), inplace="True")
df
```
