### Apply function

```python
import pandas as pd

df = pd.read_csv("/home/tanmoy/Desktop/Phitron/Practice/Module12_pandas/new_dataset.csv")
```

```python
# removing null from total marks
df.fillna({"Total Marks": df["Total Marks"].mean()}, inplace=True)
df["Total Marks"] = df["Total Marks"].round(1)
df
```

```python
# Min Max scaling
# min_max_scaling = (x - mn) / (mx - mn)

mn = df["Total Marks"].min()
mx = df["Total Marks"].max()

df["Scaled Marks"] = df["Total Marks"].apply(lambda x : (x - mn) / (mx - mn))
df
```

```python
# Grading using apply 

def grading(marks):
    if marks > 260:
        return 'A+'
    elif marks > 250:
        return 'A'
    else:
        return 'A-'
    

df["Grade"] = df["Total Marks"].apply(grading)

df
```

```python
# total marks

def total_marks(df):
    a = df["Data Structure Marks"] * 3
    b = df["Python Marks"] * 2
    c = df["Algorithm Marks"] * 5

    return a + b + c

df["New Total"] = df.apply(total_marks, axis=1)
df

```
