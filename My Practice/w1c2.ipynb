```python
def adder(a, b):
    """
    Add two number 
    a -> int 
    b -> int
    """
    sum = a + b
    return sum
print(adder(10,20) * 5)
help(adder)

```

```python
def a(*args):
    for i in args:
        print(i)

a(1,2,3,4) #multiple inpue ami eksathe print korte parteci
```

```python
def b(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

b(a = 5, b = 10, name = "tanmoy")
```

Generator

```python
l = [x for x in range (500)]

def gen (chunk_size, l):
    for i in range(len(l)):
        yield l[i: i + chunk_size]

x = gen(5, l)

print(type(x))

print(next(x))
print(x)

# jeikhane amr onek boro data ami seitake venge venge niye kaje lagacchi
# kaje lagacchi mane bola jay memory tea load kortechi
```
