```python
def introduce (name, age):
    print(f"My name is {name} and I am {age} years old.")

introduce("Rahim", 28)
```

```python
d = [x for x in range(1, 9)]

def batch_generator(data, s):
    for i in range(0, len(data), s):
        yield data[i : i + s]

g = batch_generator(d, 3)

for i in g:
    print(i)
```

```python
square = lambda x : x * x

for i in range(1, 6, 1):
    print(square(i), end= " ")
```

```python
def calculate_power(base, exponent=None):
    if exponent == None:
        exponent = 2
    
    return base**exponent

print(calculate_power(base=5, exponent=3))
print(calculate_power(5))
```

```python
a = [1,2,3,4,5,6]
a = map(lambda x : x * x, a)
a = filter(lambda x : x % 2 == 0, a)
for i in a:
    print(i)
```
