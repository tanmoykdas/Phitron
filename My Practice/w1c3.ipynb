## Right approach that handels file.close() by default

```python
with open("/home/tanmoy/Desktop/Phitron/My Practice/sample.txt", "r") as file:
    d = file.read()
    print(d)
    
print(file.closed)
```

## using for loop 

```python
with open("/home/tanmoy/Desktop/Phitron/My Practice/sample.txt", "r") as file:
    for i in file:
        print(i)
```

## File pointer tell(), seek(), read()

```python
with open("/home/tanmoy/Desktop/Phitron/My Practice/hello.txt", 'r') as file:
    print(file.tell())
    print(file.read(5)) 
    print(file.read())
    print(file.tell())
    print(file.read(5))

    #tell amader coursor koi ache bole
    #read ekdom index obdi niye jay coursor sudhu samne jay 

    #ekhane last er print kaj kortiche nah karon coursor ekdom sesh e tai 

    file.seek(0)
    print(file.tell())

    print(file.read(5)) #seek diye amra coursor jeikhane kushi nite parie
```
