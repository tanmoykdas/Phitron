### Class and Object

```python
class Phone:
    category = "Electronics"

    #Constructor
    def __init__ (self, model, battery, camera):
        self.model = model
        self.battery = battery
        self.camera = camera
```

```python
apple = Phone("Iphone", 4000, 20)
vivo = Phone("Vivo", 3000, 30)

vivo.category = "Super Electronics"

print(apple.category)
print(vivo.category)

print(apple.model)
```
