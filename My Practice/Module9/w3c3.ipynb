## Numpy array manipulation

```python
import numpy as np 
arr = np.random.randint(1,10,(10, 5))
```

```python
print(arr.ndim)
print(arr.shape)
print(arr.size)
```

```python
arr1 = arr.reshape(5,10)
print(arr1.shape)
print(arr1)
print(arr1.ndim)


arr2 = arr1.flatten()
print(arr2.ndim)
print(arr2)
```

```python
arr3 = np.ravel(arr1, order = 'F')
print(arr3)

arr3 = np.ravel(arr1, order = 'C')
print(arr3)
print()
```

```python
arr1 = np.array([[1,2,3], [4,5,6]])
arr2 = np.array([[7,8,9], [9,8,7]])

print(arr1)
print(arr2)

a = np.concatenate((arr1,arr2), axis = 0) #row wise se kaj korche
print(a)

b = np.concatenate((arr1,arr2), axis = 1) #sea column wise kaj korche
print(b)
```

## Transpose matrix

```python
arr1 = np.array([[1,2,3,4,5], [5,6,7,8,9]])

print(arr1)

arr2 = arr1.T

print(arr2)
```

## Saffer way of array spliting

## Mathematical terms of Numpy


```python
a = np.array([10,20,30,40,50])
b = np.array([1,2,3,4,5])

c = a + b
print(c)

c = a - b
print(c)

c = a * b
print(c)

c = a / b
print(c)
```

## Geomentry

```python
c = np.sin(a)
print(c)

c = np.cos(a)
print(c)

c = np.rad2deg(c)
print(c)
```

## numpy sorting 

```python
mat = [[1,22,3,4,1], [6,7,8,9,0]]

x = np.sort(mat)
print(x)
print()

# horizontal sort axis = 1
x = np.sort(mat, axis=1) #sort left to right
print(x)
print()

#vertical sort axis = 0
x = np.sort(mat, axis=0) #sort top to bottom
print(x)
print()
```

## Nupy Searching

```python
a = np.array([1,2,3,4,5])

b = np.where(a == 2)
print(b)

b = np.where(a > 2, 0, -1)
print(b)
b = np.where(a > 2)
print(b)
```

## to see where max and min value 

```python
a = np.array([1, 5, 2, 3, 4])

x = np.argmax(a)
print(x)

x = np.argmin(a)
print(x)
```

## Count elements

```python
a = np.random.randint(1, 100, size = 100,)
print(a)

x = np.count_nonzero(a > 20)
print(x)

y = np.unique(a, return_counts = True )

x,y = np.unique(a, return_counts = True )
print(x)
print(y)
```

## Statistical function

```python
x = np.genfromtxt('student_scores.csv', delimiter = ',', skip_header=True)
print(x)

first_clm = x[:,0]
print(first_clm)
mn = np.min(first_clm)
print(mn)

mx = np.max(first_clm)
print(mx)

mean = np.mean(first_clm)
print(mean)

median = np.median(first_clm)
print(median)

scatter = np.std(first_clm)
print(scatter)
```
