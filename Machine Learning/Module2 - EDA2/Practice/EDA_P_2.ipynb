### Multivariant analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/home/tanmoy/Desktop/Phitron/Machine Learning/Module2 - EDA2/Practice/titanic_data_updated.csv")

df.sample(5)
```

### Bivariant analysis

```python
grouped_data = df.groupby('Sex')['Survived'].value_counts()
grouped_data
```

### Bivariant plot

```python
sns.countplot(data = df, x='Sex', hue='Survived')
plt.show()
```

```python
grouped_data = (df.groupby('Sex')['Survived'].value_counts() / len(df)) * 100
grouped_data
```

```python
grouped_data = df.groupby('Sex')['Survived'].value_counts(normalize=True)
grouped_data
```

### Abother analysis with Pclass

```python
sns.countplot(data = df, x = 'Pclass', hue='Survived')
plt.show()
```

```python
survival_rate = df.groupby('Pclass')['Survived'].value_counts()
survival_rate
```

```python
survival_rate = df.groupby('Pclass')['Survived'].value_counts(normalize=True)
survival_rate
```
