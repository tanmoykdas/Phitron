**4. নিচের code টি একটি Countplot তৈরি করবে যেখানে passenger class অনুযায়ী count দেখাবে এবং Gender দিয়ে আলাদা করবে।**

*Hint: x = ticket class column, hue = gender column*

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("titanic")

sns.countplot(data=df, x='class', hue='sex')

plt.title("Passenger Class by Gender")
plt.show()
```

**5. নিচের code টি City অনুযায়ী average Salary দেখাবে।**

*Hint: Cat-Num relationship দেখানোর plot, x = category, y = number*

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({
    "City": ["Dhaka", "Dhaka", "CTG", "CTG", "Sylhet"],
    "Salary": [50000, 60000, 45000, 55000, 40000]
})

sns.barplot(data=df, x='City', y='Salary')

plt.title("Average Salary by City")
plt.show()
```

**6. নিচের code টি Age column এর smooth distribution curve দেখাবে।**

*Hint: Histogram এর smooth version যে plot দেখায়*

```python
import seaborn as sns
import matplotlib.pyplot as plt

df = sns.load_dataset("titanic")

sns.kdeplot(data=df, x='age')

plt.title("Age Distribution")
plt.show()
```

**7. নিচের code টি Height এবং Weight এর মধ্যে সম্পর্ক দেখাবে।**

*Hint: দুটো numerical variable এর মধ্যে relationship দেখানোর plot*

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

df = pd.DataFrame({
    "Height": [150, 160, 170, 175, 180],
    "Weight": [50, 60, 70, 75, 80]
})

sns.scatterplot(data=df, x='Height', y='Weight')

plt.title("Height vs Weight")
plt.show()
```

**8. নিচের code টি একটি dataset এর automatic EDA report তৈরি করবে।**
