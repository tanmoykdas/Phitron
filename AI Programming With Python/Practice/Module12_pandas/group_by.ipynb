```python
import pandas as pd

df = pd.read_csv("/home/tanmoy/Desktop/Phitron/Practice/Module12_pandas/student_data.csv")
```

```python
df
```

```python
def total_marks(df):
    return df["Data Structure Marks"] + df["Algorithm Marks"] + df["Python Marks"]

df["Total Marks"] = df.apply(total_marks, axis=1)

df
```

```python
instructor_group = df.groupby("Instructor")

instructor_group
```

```python
for i_name, data in instructor_group:
    print(i_name)
    display(data)
```

```python
mean_marks = instructor_group["Total Marks"].mean()

df["Total Marks"] = df["Total Marks"].fillna(df["Instructor"].map(mean_marks))
```
