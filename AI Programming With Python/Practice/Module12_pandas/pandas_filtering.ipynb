```python
import pandas as pd
import numpy as np

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace", "Hannah","Sakib"],
    "City": ["New York", "Los Angeles", "Newark", "Boston", "New Delhi", "Chicago", "New Orleans", "Houston","H Los Ang"],
    "Department": ["HR", "IT", "Finance", "IT", "HR", "Marketing", "Finance", "HR", "HR"],
    "Salary": [50000, 60000, 55000, 70000, 52000, 58000, 62000, 51000,70000]
}

df = pd.DataFrame(data)

df
```

### Filtering

```python
d = df.loc[df['City'].str.contains("New")]

d
```

```python
d = df.loc[df['City'].str.contains("new")]

d

# kichu show kore nah karon eita case sensitive
```

```python
# for case sensitive

d = df.loc[df['City'].str.contains("new", case=False)]

d
```

```python
# Staring e ami khujbo 

d = df.loc[df['City'].str.contains(r"^Delhi")]
d
# ^ sign diyei muloto eita je surute thakte hobe oita mean kore
# r diye bojhay regular expression
```

```python
# amra ekta list o pass korte pari

d = df.loc[df['City'].str.contains(r"^['h', 'n', 'b']", case=False)]
d

# case sensitive false korchi ar list cekch dice first index city er
```

```python
# $ sign diye last e ache kina check kora jay

d = df.loc[df['City'].str.contains(r"rk$", case=False)]
d
```
