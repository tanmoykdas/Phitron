```python
import pandas as pd

data = {
    "Name": ["Alice", "Bob", "Charlie", "David", "Eve", "Frank", "Grace", "Hannah","Sakib"],
    "City": ["New York", "Los Angeles", "Newark", "Boston", "New Delhi", "Chicago", "New Orleans", "Houston","H Los Ang"],
    "Department": ["HR", "IT", "Finance", "IT", "HR", "Marketing", "Finance", "HR", "HR"],
    "Salary": [50000, 60000, 55000, 70000, 52000, 58000, 62000, 51000,70000]
}

df = pd.DataFrame(data)
```

```python
df
```

### Handling duplicates

```python
df["Department"].duplicated()
```

```python
df["Department"].duplicated().sum() # koyta duplicate ache dekhay
```

```python
df.drop_duplicates() # eita mainly index based duplicate remove kore
```

```python
df.drop_duplicates(subset="Department") 
# ekta column ba multiple column er er theke remove kore
```

```python
df.drop_duplicates(subset="Department", keep="first") 
# first er index take rakhe
```

```python
df.drop_duplicates(subset="Department", keep="last") 
# last er index take rakhe
```

```python
df # that didnt change caude it creates a copy 
```

```python
df.drop_duplicates(subset="Department", keep="last", inplace=True) 
```

```python
df # inplace true korar karone eta permanent change hoice
```
