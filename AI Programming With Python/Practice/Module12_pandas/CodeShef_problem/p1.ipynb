```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [5, np.nan, np.nan, None, 10],
    'C': ['a', 'b', 'c', None, 'e']
})
                   
print(df)  

a = df.fillna({'A': 0, 'B': 99, 'C': 'Unknown'})
print(a)

b = df.copy()

b['B'] = b['B'].ffill()
print(b)

df['A'] = df['A'].fillna(df['A'].mean())
df['B'] = df['B'].fillna(df['B'].mean())
df['C'] = df['C'].fillna("Unknown")

print(df)
```
