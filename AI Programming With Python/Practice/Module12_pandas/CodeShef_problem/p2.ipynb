```python
import pandas as pd
import numpy as np

df = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [5, np.nan, np.nan, None, 10],
    'C': ['a', 'b', 'c', None, 'e']
})
                   
print(df)  

a = df.copy()
a = a.dropna(subset="A")
print(a)


a = a.dropna(subset=["B", "C"], how="all")
print(a)

df = df.dropna(subset=["A","B"], how="all")
print(df)
```
