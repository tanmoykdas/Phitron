```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("/home/tanmoy/Desktop/Phitron/Practice/Module14_Seaborn/sns_data.csv")
df.head(10)
```

### Line plot using sea born

```python
# axes level 

sns.lineplot(data=df, x='study_hours', y='test_score', errorbar=None, hue='gender')
plt.show()
```

```python
# figure level 

sns.relplot(kind='line', data=df, x='study_hours', y='test_score', errorbar=None, hue='gender')
plt.show()
```

```python
sns.relplot(kind='line', data=df, x='study_hours', y='test_score', errorbar=None, hue='subject')
plt.show()
```

### Scatter plot using seaborn

```python
sns.scatterplot(data=df, x='study_hours', y='test_score', hue='gender', style='subject', size='Tshirt_size')
plt.show()
```

```python
sns.relplot(kind='scatter', data=df, x='study_hours', y='test_score', hue='gender', style='subject', size='Tshirt_size')
plt.show()
```

### Facet Plot Using Seaborn

```python
student = pd.read_csv("/home/tanmoy/Desktop/Phitron/Practice/Module14_Seaborn/student_dataset_complete.csv")
student.head()
```

```python
sns.relplot(kind='scatter', data=student, x='study_hours', y='test_score', col='gender', row='hostel')
plt.show()
```

```python
sns.relplot(kind='scatter', data=student, x='study_hours', y='test_score', col='gender', row='class_name', style='gender', color='red', s=80)
plt.show()
```

### Histogram using Seaborn

```python
# figure level
sns.displot(data=student, x='attendance_rate', color='red', bins=15, col='gender', row='hostel')
plt.show()
```

### KDE plot using seaborn 

```python
sns.kdeplot(data=student, x='attendance_rate', hue='gender')
plt.show()
```

### Count plot and bar plot

```python
sns.countplot(data=student, x='gender')
plt.show()
```

```python
sns.barplot(data=student, x='gender', y='Marks', errorbar=None, estimator='mean')
plt.show()
```

### regplot using seaborn

```python
sns.regplot(data=student, x='study_hours', y='test_score')
plt.show()
```

```python
sns.lmplot(data=student, x='study_hours', y='test_score', hue='gender')
plt.show()
```

### Pair plot and Joint plot

```python
student.head(4)
```

```python
dat = student[["attendance_rate","study_hours","test_score"]]

sns.pairplot(data=dat)
plt.show()
```

```python
sns.jointplot(data=dat, x='study_hours', y='test_score')
plt.show()
```
