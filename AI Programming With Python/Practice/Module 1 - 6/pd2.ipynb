### given a file name x.txt

### count total character, number of line, number of word

### save it to y.txt

```python
with open("/home/tanmoy/Desktop/Phitron/My Practice/x.txt", 'r') as file:
    text = file.readlines()
    word_c = 0
    character_c = 0
    for i in range(len(text)):
        count = 1
        for j in range(len(text[i])):
            character_c += 1
            if text[i][j] == " ":
                count += 1
        word_c += count
        
    with open("/home/tanmoy/Desktop/Phitron/My Practice/y.txt", 'w') as file:
        file.write(f"x.txt File has {len(text)} line.\n")
        file.write(f"x.txt file has {word_c} word and {character_c} character.")
        
```

```python
def read_from_position(filename, position):
    with open(filename, 'r') as file:
        file.seek(position)
        c = file.read()
        print(c)

read_from_position("/home/tanmoy/Desktop/Phitron/My Practice/hello.txt", 7)
```

```python
def safe_read(filename):
    with open(filename, 'r') as file:
        try:
            print(file.read())
        except Exception as e:
            print(e)
        finally:
            print("Operation succesfull!")

try :
    safe_read("/home/tanmoy/Desktop/Phitron/My Practice/missing.txt")
except Exception as e:
    print(e)
else:
    with open("/home/tanmoy/Desktop/Phitron/My Practice/missing.txt", 'w') as file:
        file.write("Hello world!")
finally:
    print("Operation, successfull!")
```
